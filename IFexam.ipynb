```python
# if문은 제어문으로 참, 거짓을 판단하여 흐름을 조절함

# ex : 만약에 돈이 있으면 택시를 타고가고, 없으면 걸어가라!!!
#       if        True                    False

# if 문법 암기 :
# if 변수(참/거짓) :
# 탭(띄어쓰기) 실행문 - > 참일 때 실행문
# 탭(띄어쓰기) 실행문 - > 참일 때 실행문
# else : -> 거짓일 때 실행문을 만듦
# 탭(띄어쓰기) 실행문 - > 참일 때 실행문
# 탭(띄어쓰기) 실행문 - > 참일 때 실행문

money = False    # 돈이 있다.
if money :      # 만약에 money값 안에 True이면 아래 들여쓰기 한 곳을 실행한다.
    print("택시를 타고 갑니다.")    # True일 때 출력문
else :          # 위에 if문에 False가 들어올 때 아래 들여쓰기 한 곳 실행문
    print("걸어가라!!!")
    print("학교에 도착하였습니다.")
```

```python
clock = int(input("현재 몇시인가요?"))
if clock >= 22 :
    print("현재 %d시 입니다." % clock)
    print("잠을 잡니다.")
else :
    print("현재 %d시 입니다." % clock)
    print("게임을 합니다.")
```

```python
money = int(input("현재 얼마를 가지고 있나요?"))
if money >= 30000 :
    print("현재 %d원을 가지고 있습니다." % money)
    print("구매할 수 있습니다.")
else :
    print("현재 %d원을 가지고 있습니다." % money)
    print("구매할 수 없습니다.")
```
